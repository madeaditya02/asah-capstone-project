## **Univariate Analysis**

### **Numerical Features Distribution**

#### Overview:
Dataset memiliki **10 variabel numerik** yang terbagi menjadi:
- **Customer characteristics** (3): age, duration, campaign
- **Contact history** (2): pdays, previous  
- **Economic indicators** (5): emp.var.rate, cons.price.idx, cons.conf.idx, euribor3m, nr.employed (diukur per periode waktu, bukan data unik untuk tiap nasabah)**bold text**

#### General Observation:
Mayoritas variabel menunjukkan **right-skewed distribution** dengan outliers. Variabel ekonomi menunjukkan pola **multimodal** yang mencerminkan campaign dilakukan dalam periode waktu berbeda.

---

#### Interpretasi per Variable:

**age**: Distribusi normal, mayoritas 30-40 tahun.

**duration**: Right-skewed, mayoritas panggilan < 5 menit (median: 180s).
Terdapat extreme outliers hingga >4000 detik (67 menit).

**campaign**: Right-skewed, mayoritas hanya 1-2 kontak (75% â‰¤3 kontak).
Terdapat outliers >20 kontak yang mengindikasikan follow up yang sebanyak itu (>20 kali).

**pdays**: Distribusi bimodal dengan spike ekstrem di 999.
**Important note**: 999 adalah kode khusus yang berarti "never contacted before",
bukan nilai 999 hari yang sebenarnya. 96.3% data memiliki nilai ini.

**previous**: Right-skewed, mayoritas 0 (86% customers tidak pernah di-follow up). Hanya sebagian kecil customers yang pernah dihubungi di campaign sebelumnya.

**emp.var.rate (Employment Variation Rate)**:
Tingkat perubahan employment rate (indikator pertumbuhan ekonomi).
Distribusi multimodal dengan 3-4 peaks. Range: -3.4 hingga +1.4.
- Negative values mengindikasikan ekonomi kontraksi/resesi
- Positive values mengindikasikan ekonomi ekspansi/pertumbuhan
- Mayoritas campaign dilakukan saat ekonomi dalam kondisi sulit (negative values)

**cons.price.idx (Consumer Price Index)**:
Indeks harga konsumen (ukuran inflasi). Range: 92.2-94.8.
Terdapat 3-4 clusters yang jelas (~92.5, ~93.0, ~93.8, ~94.0).
Narrow range (2.6 points) mengindikasikan periode inflasi relatif stabil.
Setiap cluster kemungkinan mewakili campaign period berbeda (quarterly/semester).

**cons.conf.idx (Consumer Confidence Index)**:
Indeks kepercayaan konsumen (sentimen ekonomi masyarakat). Range: -50.8 hingga -26.9.
Distribusi trimodal dengan 3 peaks jelas (~-47, ~-42, ~-36).
**All negative values** mengindikasikan consumer sentiment buruk sepanjang periode campaign.
Consumer confidence yang lebih tinggi (mendekati -26) berpotensi meningkatkan conversion rate.

**euribor3m (Euribor 3-month Rate)**:
Euro Interbank Offered Rate untuk 3 bulan (benchmark interest rate). Range: 0.6%-5.0%.
Distribusi bimodal dengan 2 clusters sangat jelas (~1% dan ~5%).
Mencerminkan 2 era ekonomi sangat berbeda:
- Low rate period (~1%): deposit products kurang menarik (returns rendah)
- High rate period (~5%): deposit products lebih menarik (returns tinggi)
Directly affects product value proposition.

**nr.employed (Number of Employees)**:
Jumlah karyawan dalam ribuan (ukuran employment level nasional). Range: 4,963-5,228k.
Distribusi bimodal dengan 2 peaks (~5,100 dan ~5,200).
Mencerminkan 2 employment regimes: ekonomi lemah vs recovery phase.

## **Outlier Detection**

**duration**: Boxplot 'yes' memiliki median (~ 400 detik) dan IQR yang jauh lebih tinggi dibandingkan boxplot 'no' (~ 150 detik). Walaupun nilai dari duration diketahui setelah panggilan dilakukan, jadi ga bisa digunakan untuk targetting awal.

**euriborn3m**: Kelompok 'yes' memiliki median yang sangat rendah (~ 1.2%), sedangkan kelompok 'no' memiliki median yang sangat tinggi (~ 4.8%). Hal ini menunjukkan kalau keberhasilan konversi terkonsentrasi pada periode di mana suku bunga antar bank sedang rendah.

**nr.employed**: Kelompok 'yes' memiliki median yang jauh lebih rendah (~ 5099k) dibandingkan kelompok 'no' (~ 5228k). Konversi lebih banyak terjadi pada periode di mana tingkat penyerapan tenaga kerja lebih rendah (potensi resesi atau pemulihan awal).

**emp.var.rate**: Kelompok 'yes' memiliki median dan IQR yang sepenuhnya negatif (~ -1.8), sedangkan kelompok 'no' memiliki median positif (~ +1.1).

**cons.conf.idx**: Kelompok 'yes' memiliki median yang sedikit lebih tinggi (kurang negatif, ~ -38) dibandingkan kelompok 'no' (~ -41). Hal ini menunjukkan walaupun kondisi ekonomi sedang tidak baik, tetapi konversi lebih mungkin terjadi ketika kepercayaan konsumen sedikit membaik.

**cons.price.idx**: Kelompok 'yes' memiliki median yang sedikit lebih rendah (~ 93.2) dibandingkan 'no' (~ 93.8), yang berarti konversi lebih sering terjadi pada periode dengan inflasi sedikit lebih rendah.

**campaign**: Nasabah yang berlangganan (yes) cenderung dihubungi lebih sedikit. Outlier dengan jumlah kontak ekstrim (>20) hampir secara eksklusif berada di kelompok 'no', yang mengindikasikan adanya campaign fatigue (menghubungi terlalu sering menurunkan peluang sukses).

**age**: Boxplot 'yes' dan 'no' hampir identik.

```python
outlier_summary = {}
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    outlier_pct = len(outliers) / len(df) * 100
    outlier_summary[col] = {'count': len(outliers), 'percentage': outlier_pct}
    print(f"{col}: {len(outliers)} outliers ({outlier_pct:.2f}%)")
```

**1. Multikolinearitas Ekstrem pada Indikator Ekonomi**

Ketika satu indikator (seperti suku bunga) naik, indikator lainnya (seperti jumlah tenaga kerja) juga naik. Walaupun kalau model yang dipakai itu kayak XGBoost atau Random Forest, bisa lebih sederhana.

Oleh karena itu, tidak boleh untuk memasukkan semua fitur ini secara bersamaan ke dalam model. Bisa untuk memilih salah satu fitur atau menggunakan PCA (untuk reduksi dimensi).

**3. Konfirmasi "Gelombang" Kampanye**

Fitur-fitur seperti:

`previous` dan `emp.var.rate`: -0.42

`previous` dan `euribor3m`: -0.45

`previous` dan `nr.employed`: -0.5

Fitur `previous` berkorelasi negatif dengan semua indikator ekonomi, yang berarti:

1. Nasabah dengan previous = 0 (kontak baru di kampanye ini) dihubungi saat indikator ekonomi tinggi (nilai emp.var.rate, euribor3m, nr.employed tinggi).

2. Nasabah dengan previous > 0 (dihubungi di kampanye sebelumnya) dihubungi saat indikator ekonomi rendah.

**4. Prediktor Independen**

Fitur `age`, `duration`, dan `campaign` memiliki korelasi yang sangat lemah dengan hampir semua fitur lainnya. Hal ini berarti fitur-fitur ini memberikan informasi unik yang tidka tumpang tindih dengan fitur lain.

### **Categorical Features**

```python
from scipy.stats import chi2_contingency

for col in categorical_cols[:-1]:  # exclude 'y'
    # Stacked bar
    pd.crosstab(df[col], df['y'], normalize='index').plot(kind='bar', stacked=True)
    plt.show()

    # Chi-square test
    chi2, p_value, _, _ = chi2_contingency(pd.crosstab(df[col], df['y']))
    print(f"{col}: p-value = {p_value:.4f}")
    print("\n")
```

**Interpretasi p-value pada chi-square**

**1. Prediktor kuat**

`job`: p-value = 0.0000 (perbedaan untuk 'student' dan 'retired')

`contact`: p-value = 0.0000 (perbedaan bahwa 'cellular' > 'telephone')

`month`: p-value = 0.0000 (perbedaan di bulan mar, sep, oct, dec sangat tinggi)

`poutcome`: p-value = 0.0000 (perbedaaan bahwa kategori 'success' > 60% say 'yes')

**2. Prediktor lemah**

`marital`: p-value = 0.0000 (memang nilai p-value kecil, tetapi chart-nya rata, sehingga perbedaaannya terlalu kecil untuk dipertimbangkan)

`education`: p-value = 0.0000 (perbedaan tipis)

`default`: p-value = 0.0000 (perbedaan tipis)

`day_of_week`: p-value = 0.0000 (perbedaan tipis)

**Tidak ada hubungan**

`housing`: p-value = 0.0583 (p-value > 0.05, sehingga tidak ada hubungan yang signifikan)

`loan`: p-value = 0.5787 (p-value sangat besar, jadi tidak ada hubungan)

## **Feature Engineering Exploration/Summary**

```python
df['age_group'] = pd.cut(df['age'], bins=[0, 30, 40, 50, 60, 100],
                          labels=['<30', '30-40', '40-50', '50-60', '60+'])

plt.figure(figsize=(10, 6))
pd.crosstab(df['age_group'], df['y'], normalize='index').plot(kind='bar')
plt.title('Conversion Rate by Age Group')
plt.ylabel('Proportion')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

print(pd.crosstab(df['age_group'], df['y'], normalize='index'))
```

```python
df['campaign_category'] = pd.cut(df['campaign'],
                                  bins=[0, 1, 3, 5, float('inf')],
                                  labels=['Single', 'Few (2-3)', 'Many (4-5)', 'Excessive (6+)'])

plt.figure(figsize=(10, 6))
pd.crosstab(df['campaign_category'], df['y'], normalize='index').plot(kind='bar')
plt.title('Conversion Rate by Campaign Intensity')
plt.ylabel('Proportion')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

print("\n=== Campaign Category Distribution ===")
print(df['campaign_category'].value_counts())
print("\n=== Conversion Rate by Campaign Category ===")
print(pd.crosstab(df['campaign_category'], df['y'], normalize='index'))
```

```python
df['had_previous_success'] = (df['poutcome'] == 'success').astype(int)

plt.figure(figsize=(8, 6))
pd.crosstab(df['had_previous_success'], df['y'], normalize='index').plot(kind='bar')
plt.title('Conversion Rate: Previous Success vs No Previous Success')
plt.ylabel('Proportion')
plt.xlabel('Had Previous Success')
plt.xticks(ticks=[0, 1], rotation=0, labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

```python
df['call_duration_category'] = pd.cut(df['duration'],
                                       bins=[0, 120, 300, 600, float('inf')],
                                       labels=['Very Short (<2min)', 'Short (2-5min)',
                                               'Medium (5-10min)', 'Long (>10min)'])

plt.figure(figsize=(10, 6))
pd.crosstab(df['call_duration_category'], df['y'], normalize='index').plot(kind='bar')
plt.title('Conversion Rate by Call Duration')
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
median_euribor = df['euribor3m'].median()
median_employed = df['nr.employed'].median()

df['good_economic_condition'] = ((df['euribor3m'] < median_euribor) &
                                  (df['nr.employed'] > median_employed)).astype(int)

plt.figure(figsize=(8, 6))
pd.crosstab(df['good_economic_condition'], df['y'], normalize='index').plot(kind='bar')
plt.title('Conversion Rate by Economic Condition')
plt.ylabel('Proportion')
plt.xlabel('Good Economic Condition')
plt.xticks(ticks=[0, 1], rotation=0, labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

```python
economic_vars = ['emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
plt.figure(figsize=(8, 6))
sns.heatmap(df[economic_vars].corr(), annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation: Economic Variables')
plt.tight_layout()
plt.show()
```
